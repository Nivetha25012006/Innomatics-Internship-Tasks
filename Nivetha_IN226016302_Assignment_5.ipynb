**Problem 1: Social Media – Post Engagement Analyzer**

```python
def post_engagement(likes):
    total_likes = 0
    for like in likes:
        total_likes += like

    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)

post_engagement([200, 300, 250, 300])
```

**Problem 3: Agriculture – Rainfall Adequacy Checker**

```python
def rainfall_checker(rainfall, required):
    total = 0
    for r in rainfall:
        total += r

    average = total / len(rainfall)
    status = "Adequate Rainfall" if average >= required else "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

rainfall_checker([70, 75, 72, 71], 70)
```

**Problem 4: Social Media – Duplicate Account Detection**

```python
def duplicate_accounts(usernames):
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

duplicate_accounts(["user1", "user2", "user1"])
```

**Problem 6: Agriculture – Premium Crop Price Filter**

```python
def premium_crops(prices):
    premium = []
    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)

premium_crops([1500, 2500, 1800, 3200])
```

**Problem 7: System Monitoring – Application Health Checker**

```python
def system_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)

system_health(7)
```

**Problem 8: Banking – Daily Transaction Limit Checker**

```python
def transaction_check(amount):
    limit = 50000
    status = "Approved" if amount <= limit else "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

transaction_check(60000)
```

**Problem 9: E-Learning – Student Attendance Eligibility System**

```python
def attendance_eligibility(attendance):
    total_classes = len(attendance)
    attended = 0

    for day in attendance:
        if day == 1:
            attended += 1

    percentage = (attended / total_classes) * 100
    status = "Eligible" if percentage >= 75 else "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

attendance_eligibility([1,1,1,0,1,1,1,1,0,1])
```

**Problem 10: Smart Electricity Bill Analyzer**

```python
def electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage

amount, status = electricity_bill(250)
print("Total Bill:", amount)
print("Usage Level:", status)
```
