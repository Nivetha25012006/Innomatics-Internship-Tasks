**Problem 1: Mobile Recharge Validation System**

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    if amount >= 50 and amount in valid_plans:
        return True
    return False

while True:
    amt = int(input("Enter recharge amount: "))
    if validate_recharge(amt):
        print("Recharge Successful ✅")
        break
    else:
        print("Invalid recharge amount. Try again ❌")
```

**Problem 2: Inventory Reorder Alert System**


```python
def check_inventory(stock):
    for product, quantity in stock.items():
        if quantity < 15:
            print(product, ": Reorder Alert")
        else:
            print(product, ": Stock OK")

inventory = {
    "Rice": 10,
    "Sugar": 20,
    "Oil": 8,
    "Wheat": 25
}

check_inventory(inventory)
```

**Problem 3: Student Result Processing System**

```python
def student_result(marks):
    total = 0
    for m in marks:
        total += m
    avg = total / len(marks)

    if avg >= 50:
        return "Pass"
    else:
        return "Fail"

marks_list = [60, 55, 40, 70]
result = student_result(marks_list)
print("Result:", result)
```

**Problem 4: Cab Fare Estimator with Retry Option**

```python
def calculate_fare(distance, peak):
    fare = 50 + (distance * 12)
    if peak == "yes":
        fare += fare * 0.25
    return fare

while True:
    km = float(input("Enter distance (km): "))
    peak_hour = input("Peak hour? (yes/no): ").lower()

    total_fare = calculate_fare(km, peak_hour)
    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to retry? (yes/no): ").lower()
    if retry != "yes":
        break
```

**Problem 5: Employee Attendance Eligibility Checker**

```python
def attendance_check(attendance):
    present = 0
    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance_list = ["P", "P", "A", "P", "P", "A", "P"]
status = attendance_check(attendance_list)
print("Attendance Status:", status)
```

**Problem 6: Password Strength Checker**

```python
def password_strength(password):
    if len(password) < 8:
        return "Weak Password"

    has_digit = False
    has_special = False

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

pwd = input("Enter password: ")
print(password_strength(pwd))
```
