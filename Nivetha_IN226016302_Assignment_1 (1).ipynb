**Task 1: Conceptual Understanding**

**1. Difference between "Love" and "love" in NLP**
In NLP, "Love" and "love" are treated as different tokens if case sensitivity is not handled.
This can increase vocabulary size unnecessarily.
After preprocessing (lowercasing), both become the same → "love"
Example:
Without preprocessing: ["Love", "love"] → 2 tokens
With preprocessing: ["love", "love"] → 1 token

**2. What happens if stopwords are not removed?**
Text will contain many irrelevant words like "is", "the", "and"
This:
    Increases data size
    Reduces model efficiency
    Adds noise
Machine learning models may focus on unimportant words

**3. When removing stopwords can be harmful **

Scenario 1: Sentiment Analysis

Sentence: "I am not happy"
If "not" is removed → "I am happy"  (wrong meaning)

Scenario 2: Question Answering

Sentence: "What is your name?"
Removing stopwords → "name" (loses intent)

**4. Stemming vs Lemmatization**
Feature	       Stemming	              Lemmatization
Method      	Cuts words	           Uses dictionary
Output	   Not always meaningful    	Meaningful word
Example	    "running" → "run"	       "running" → "run"
Example    	"better" → "better"     	"better" → "good"

 Conclusion: Lemmatization is more accurate

**Task 2: Advanced Preprocessing Function**

```python
import re

def preprocess_text(text):
    if not text or text.strip() == "":
        return [], ""

    # Lowercase
    text = text.lower()

    # Remove URLs and emails
    text = re.sub(r'http\S+|www\S+|@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove repeated characters (soooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove special characters & emojis
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenize
    tokens = text.split()

    # Remove short words (except "no", "not")
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

**Task 4: Token Analytics**

```python
def analyze_tokens(tokens):
    total = len(tokens)
    unique = len(set(tokens))
    avg_length = sum(len(word) for word in tokens) / total if total > 0 else 0
    return total, unique, avg_length
```

**Task 5: Frequency Analysis**

```python
from collections import Counter

all_tokens = []

for sentence in sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 frequent:", counter.most_common(10))
print("Least frequent:", counter.most_common()[-5:])
```

**Task 6: Full Pipeline**

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```
