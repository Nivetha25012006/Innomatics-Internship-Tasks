**Problem 1: Smart Light Controller**

```python
class SmartLight:
    def __init__(self, name, action):
        self.name = name
        self.action = action

    def show_status(self):
        print(self.name, "is", self.action)


light = SmartLight("Bedroom Light", "ON")
light.show_status()
```

**Problem 2: Employee ID Card System**

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


emp = Employee("Rahul", "EMP102", "IT")
emp.display()
```

**Problem 3: Mobile Contact Record**

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def show_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


c = Contact("Anita", "9876543210")
c.show_contact()
```

**Problem 4: Product Price Tag Generator**

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def show_price(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


p = Product("Headphones", 2499)
p.show_price()
```

**Problem 5: Movie Rating Display System**

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def show_details(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


m = Movie("Inception", 4.8)
m.show_details()
```

**Problem 6: Delivery Address Manager**

```python
class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def show_details(self):
        print("Delivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)


d = Delivery("Suman", "Hyderabad")
d.show_details()
```
