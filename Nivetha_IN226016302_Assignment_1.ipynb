1. User Login Check
Problem Statement:
Given a username and password, check whether login is successful.
username = "admin"
password = "1234"



```python
# Stored credentials
username = "admin"
password = "1234"

# Taking input from user
input_username = input("Enter username: ")
input_password = input("Enter password: ")

# Checking login condition
if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

2. Pass / Fail Analyzer
Problem Statement:
Analyze student results.
marks = [45, 78, 90, 33, 60]



```python
# List of student marks
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

# Loop through marks
for m in marks:
    if m >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Final output
print("Total Students Passed:", pass_count)
print("Total Students Failed:", fail_count)
```

3. Simple Data Cleaner
Problem Statement:
Clean and standardize user names.
names = [" Alice ", "bob", " CHARLIE "]


```python
# Original list of names
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

# Cleaning process
for name in names:
    clean_name = name.strip().lower()   # remove spaces and convert to lowercase
    cleaned_names.append(clean_name)

# Output cleaned list
print("Cleaned Names List:", cleaned_names)
```

4. Message Length Analyzer
Problem Statement:
Analyze message sizes.
messages = ["Hi", "Welcome to the platform", "OK"]


```python
# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Analyze each message
for msg in messages:
    length = len(msg)
    print(f"Message: '{msg}' → Length: {length}")

    # Check if message is long
    if length > 10:
        print("⚠️ This message is longer than 10 characters")
```

5. Error Message Detector
Problem Statement:
Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]


```python
# System log data
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

# Count ERROR entries
for log in logs:
    if log == "ERROR":
        error_count += 1

# Final output
print("Total ERROR messages:", error_count)
```
